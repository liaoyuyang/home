# 跨分组对比模板

对比同一品种在不同分组下的 all_factor 差异。

```python
import warnings
warnings.filterwarnings("ignore")
import sys
sys.path.insert(0, '/home/strat_lab')

import pandas as pd
from src.data_loader import DataLoader

dl = DataLoader()
SYMBOL = 'A'
```

```python
# 加载同一品种在不同分组下的因子表
df_g1 = dl.load_all_factor('油脂油粕', SYMBOL)
df_g2 = dl.load_all_factor('农产品', SYMBOL)

print("油脂油粕列数:", len(df_g1.columns))
print("农产品列数:", len(df_g2.columns))

# 差异列
only_g1 = set(df_g1.columns) - set(df_g2.columns)
only_g2 = set(df_g2.columns) - set(df_g1.columns)
print(f"仅在油脂油粕: {len(only_g1)}")
print(f"仅在农产品: {len(only_g2)}")
```
