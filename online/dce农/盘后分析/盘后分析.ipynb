## 3. 分析单日所有品种

```python
# 分析所有品种
DATE = dates[0]
results = analyze_all_symbols(DATE)
```

```python
# 创建对比表格
comparison = pd.DataFrame([
    {
        '品种': r['symbol'],
        '名称': r['name'],
        '总盈亏': r.get('total_pnl', 0),
        '最大盈利': r.get('max_pnl', 0),
        '最大亏损': r.get('min_pnl', 0),
        '交易次数': r.get('num_trades', 0),
        '多头持仓(分钟)': r.get('long_bars', 0),
        '空头持仓(分钟)': r.get('short_bars', 0),
    }
    for r in results
])
comparison
```

## 4. 批量分析多个日期

```python
# 分析多个日期
dates_to_analyze = ['20260512', '20260513', '20260514', '20260515']

all_results = []
for date in dates_to_analyze:
    print(f'\n{"#"*60}')
    print(f'# 日期: {date}')
    print(f'{"#"*60}')
    results = analyze_all_symbols(date, save_data=False, save_plot=False)
    for r in results:
        r['date'] = date
        all_results.append(r)
```

```python
# 汇总批量分析结果
batch_summary = pd.DataFrame(all_results)
summary_table = batch_summary.pivot_table(
    values='total_pnl', 
    index='symbol', 
    columns='date', 
    aggfunc='sum'
)
summary_table['合计'] = summary_table.sum(axis=1)
summary_table = summary_table.sort_values('合计', ascending=False)
summary_table
```

## 5. 查看原始数据

```python
# 加载并查看 factors 数据
SYMBOL = 'P'
DATE = '20260515'

factors_df = merge_factors_data(SYMBOL, DATE)
if factors_df is not None:
    print(f'Factors 数据: {len(factors_df)} 行')
    print(f'列数: {len(factors_df.columns)}')
    factors_df.head()
```

```python
# 加载并查看 predictions 数据
pred_df = merge_predictions_data(SYMBOL, DATE)
if pred_df is not None:
    print(f'Predictions 数据: {len(pred_df)} 行')
    print(f'列: {list(pred_df.columns)}')
    pred_df.head()
```

```python
# 加载并查看交易状态
status_df = load_trading_status(SYMBOL, DATE)
if status_df is not None:
    print(f'交易状态: {len(status_df)} 条')
    print(f'列: {list(status_df.columns)}')
    status_df.head()
```

## 6. 汇总历史绩效

```python
# 汇总所有日期所有品种
summary = summary_all_days()
print(f'汇总记录数: {len(summary)}')
summary.head(10)
```

```python
# 按品种汇总
if len(summary) > 0:
    symbol_summary = summary.groupby('symbol').agg({
        'num_trades': 'sum',
        'final_pos': ['last', lambda x: (x != 0).sum()]
    }).reset_index()
    symbol_summary.columns = ['品种', '总交易次数', '最终持仓', '持仓天数']
    symbol_summary
```
